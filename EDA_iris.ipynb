```python
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

iris = load_iris()
```

```python
df= pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
df['target_name'] = df['target'].apply(lambda x: iris.target_names[x])

df.head()
```

```python
for column in iris.feature_names:
    plt.figure()
    sns.histplot(df, x=column, hue='target_name', kde=True, element='step')
    plt.title(f'Histogram of {column}')
    plt.tight_layout()
    plt.show()
```

```python
sns.pairplot(df, hue="target_name", diag_kind="hist")
plt.suptitle("Scatterplot Matrix of Iris Features", y=1.02)
plt.show()
```

```python
summary_stats = df.describe(include='all')

correlation_matrix = df.corr(numeric_only=True)

class_means = df.groupby('target_name').mean(numeric_only=True)

print("Summary Statistics:")
print(summary_stats)

print("\nCorrelation Matrix:")
print(correlation_matrix)

print("\nClass-wise Means:")
print(class_means)
```
