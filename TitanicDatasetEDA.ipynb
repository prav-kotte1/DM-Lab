```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
titanic = sns.load_dataset("titanic")

# basic info
titanic.head()
```

```python
# --- Cell 3: Data Overview ---
print(titanic.info())
print(titanic.describe(include="all"))
```

```python
# --- Cell 4: Univariate Analysis - Categorical ---
plt.figure(figsize=(6,4))
sns.countplot(x="sex", data=titanic)
plt.title("Distribution of Sex")
plt.show()

plt.figure(figsize=(6,4))
sns.countplot(x="class", data=titanic)
plt.title("Distribution of Passenger Class")
plt.show()
```

```python
# --- Cell 5: Univariate Analysis - Numerical ---
plt.figure(figsize=(6,4))
sns.histplot(titanic["age"].dropna(), kde=True, bins=30)
plt.title("Age Distribution")
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(y="age", data=titanic)
plt.title("Age Boxplot")
plt.show()
```

```python
# --- Cell 6: Bivariate Analysis ---
plt.figure(figsize=(6,4))
sns.countplot(x="sex", hue="survived", data=titanic)
plt.title("Survival by Sex")
plt.show()

plt.figure(figsize=(6,4))
sns.countplot(x="class", hue="survived", data=titanic)
plt.title("Survival by Passenger Class")
plt.show()
```

```python
# --- Cell 7: Correlation Heatmap ---
# select numeric columns
corr = titanic.corr(numeric_only=True)
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# --- Cell 8: Pair Plot ---
sns.pairplot(titanic[["age","fare","survived","pclass"]].dropna(), hue="survived")
plt.show()
```
